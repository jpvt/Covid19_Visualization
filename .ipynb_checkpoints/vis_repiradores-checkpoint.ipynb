```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Distribuição dos respiradores no Brasil durante a pandemia

## Quantidade de Respiradores / Mês em cada Estado

```python
report_month = pd.read_csv('data_map2.csv')
report_month.head()
```

```python
months = [4,5,6,7,8]
month_dict = {4 : 'Abril', 5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto'}

for month in months:
    df_month = report_month[report_month['MES'] == month]
    #df_month.set_index('DESTINO',inplace = True)
    
    plt.figure(figsize=(16,8))
    plt.title(f"Pedidos de Respiradores em {month_dict[month]}")
    ax = sns.barplot(x= df_month['QT_MES'], y = df_month['DESTINO'])
    ax.set(xlabel="Quantidade de Respiradores", ylabel="Estado")
    plt.show()
    
    
```

## Valor gasto com Respiradores / Mês em cada Estado

```python
for month in months:
    df_month = report_month[report_month['MES'] == month]
    #df_month.set_index('DESTINO',inplace = True)
    
    plt.figure(figsize=(16,8))
    plt.title(f"Valor gasto com Respiradores em {month_dict[month]}")
    ax = sns.barplot(x= df_month['VL_MES'], y = df_month['DESTINO'])
    ax.set(xlabel="Valor Gasto em Reais (R$)", ylabel="Estado")
    plt.show()
    print('\n\n\n\n')
```

## Diferença de preço de acordo com o fornecedor

## Diferença de preço de acordo com o Estado

```python
gb = report_month.groupby('DESTINO')

qt_tot = []
vl_tot = []
for i in range(len(report_month)):
    st = gb.get_group(report_month['DESTINO'].iloc[i])
    qt_tot.append(st.agg({'QT_MES': ['sum']}).QT_MES.values[0])
    vl_tot.append(st.agg({'VL_MES': ['sum']}).VL_MES.values[0])

report_month['QT_TOT'] = qt_tot
report_month['VL_TOT'] = vl_tot
report_month['PREÇO_MEDIO'] = (np.array(vl_tot)/np.array(qt_tot))

report = report_month.copy().drop(columns = ['QT_MES','VL_MES','MES','longitude','latitude'])
```

```python
report2 = report.sort_values(by='PREÇO_MEDIO', ascending=True)

pal = sns.diverging_palette(235, 10, l=30, n=27, center="dark")
plt.figure(figsize=(16,8))
plt.title(f"Quanto cada Estado pagou por respirador ?")
ax = sns.barplot(x= report2['PREÇO_MEDIO'], y = report2['DESTINO'], palette=pal)
ax.set(xlabel="Valor Médio de Respiradores por Estado em Reais (R$)", ylabel="Estado")
plt.show()
```
