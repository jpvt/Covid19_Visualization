```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados_dash2.csv')
df['DATA'] = pd.to_datetime(df['DATA'])
df_selec = df.drop(columns = ['FORNECEDOR','ESTADO/MUNICIPIO','TIPO','DESTINATARIO','UF','DATA DE ENTREGA'])
df_selec['MES'] = df['DATA'].dt.month
df_selec.head()
```

```python
gb = df.groupby('DESTINO')

qt_mes = []
vl_mes = []
for i in range(len(df_selec)):
    st = gb.get_group(df_selec['DESTINO'].iloc[i]).set_index('DATA')
    qt_mes.append(st[st.index.month == df_selec['MES'].iloc[i]].agg({'QUANTIDADE': ['sum']}).QUANTIDADE.values[0])
    vl_mes.append(st[st.index.month == df_selec['MES'].iloc[i]].agg({'VALOR': ['sum']}).VALOR.values[0])

df_selec['QT_MES'] = qt_mes
df_selec['VL_MES'] = vl_mes
```

```python
df_selec
```

```python
new_df = df_selec.drop(columns = ['DATA','QUANTIDADE','VALOR'])
new_df = new_df.drop_duplicates()
new_df
```

```python
new_df.to_csv('data_map2.csv', index = False)
```
